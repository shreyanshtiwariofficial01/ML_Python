```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,root_mean_squared_error,r2_score
```

```python
data=pd.read_csv('California_updated.csv')
X=data.drop('MedHouseVal',axis=1)
y=data['MedHouseVal']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1234)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)

MSE=mean_squared_error(y_test,y_pred)
RMSE=root_mean_squared_error(y_test,y_pred)
R2=r2_score(y_test,y_pred)
```

```python
print('MSE',MSE)
print('RMSE',RMSE)
print('R2',R2)
```

**Model Save**

- We can save model in two ways

    - suning pickle : .pki

    - using joblib : .joblib

**pickle**

```python
import pickle
```

- dump

    - we will save the model in .pkl format

    - it expexts the model and location

    - it is similar to file open write format

    - hear also will open the file with write bytes format

- load

   - similar way it needs only file name in read bytes format

```python
# our model variable name is : lr
# where i want to save : same directory
# on what name i want to save: california
# on what extension i want to save : ,pkl
# file name : california,pkl
# write format is : wb

file=open('dummy.txt','w')
str1='hello'
file.write(str1)
```

```python
str1='hello'
with open('dummy1.txt','w')as file:
    file.write(str1)
```

```python
with open('california.pkl','wb') as file:
    pickle.dump(lr,file)
```

```python
with open('california.pkl','rb') as file:
    loaded_model=pickle.load(file)
```
