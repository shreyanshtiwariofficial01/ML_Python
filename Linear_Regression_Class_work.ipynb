```python
california_df.columns
```

```python
# seperate input and output data 
X= california_df.drop('MedHouseVal',axis=1)
y=california_df['MedHouseVal']
```

```python
X.columns
```

```python
# 'MedInc'= bo+b1*'HouseAge'+b2*'AveRooms'+b3*'AveBedrms'.....+b7*'Longitude'
#                  |
#                 Rsq
#                  |
#                  VIF

# 'HouseAge'= bo+b1*'MedInc'+b2*'AveRooms'+b3*'AveBedrms'.....+b7*'Longitude'
#                  |
#                 Rsq
#                  |
#                  VIF

# 8 times
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
X.values
```

```python
X.columns
```

```python
# we need to pass the array of all the input column data
variance_inflation_factor(X.values,7)
#'MedInc'= bo+b1*'HouseAge'+b2*'AveRooms'+b3*'AveBedrms'.....+b7*'Longitude'
```

```python
vif=[]
for i in range(len(X.columns)):
    value=variance_inflation_factor(X.values,i)
    vif.append(value)
```

```python
pd.DataFrame(vif,
             index=X.columns,
             columns=['VIF'])
```

```python
X.columns
```

```python
final_Cols=['MedInc','HouseAge','Population','AveOccup','AveRooms','Latitude']

final_df=pd.DataFrame()
final_df=X[final_Cols].copy()
final_df['MedHouseVal']=y
final_df
```
